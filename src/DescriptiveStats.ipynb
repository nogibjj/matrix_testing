```python
import lib
import pandas as pd
```

```python
# Using common functions from lib.py
data = pd.read_csv('../data/iris_data.csv')
TARGET_COLUMN = 'sepal_length'

print('Target Column: ', TARGET_COLUMN)
print("Mean of sepal length: ", round(lib.return_mean(data, TARGET_COLUMN), 2))
print("Median of sepal length: ", round(lib.return_median(data, TARGET_COLUMN), 2))

print("25 th quantile of sepal length: ", 
      round(lib.return_25th_quantile(data, TARGET_COLUMN), 2))

print("Standard Deviation of sepal length: ", 
      round(lib.return_std_dev(data, TARGET_COLUMN), 2))
```
